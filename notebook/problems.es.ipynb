## Ejercicio 1 

Dos dados se lanzan una vez y se observa el total obtenido. Usa una simulación para encontrar la probabilidad estimada de que la puntuación sea mayor a 7 o un número par. Una simulación es una repetición del mismo experimento multiples veces para observar su comportamiento:

- Ejecuta el experimento 1000 veces (lanza 2 dados 1000 veces, y suma el número de ambos dados).
- Lleva ua cuenta de los números y las veces que la suma fue mayor a 7 o un número par.
- Divide el número del paso 2 entre el número de iteraciones(1000).

```python
import random
import math
```

```python
# Simulacion: repetición del mismo experimento multiples veces para observar su comportamiento.
# Evento: nro > 7 ó nro par

# Funcion para lanzar el dado
def lanzar_dado():
    return random.randint(1,6)

# Funcion para la simulacion
def experimento(nro):
    evento = 0
    for i in range(nro):
        resul1 =  lanzar_dado()
        resul2 = lanzar_dado()
        total = resul1 + resul2
        if total > 7 or total % 2 == 0:
            evento += 1
    return evento
veces = 1000
evento = experimento(veces)
print(f"El resultado de lanzar dos dados {veces} veces, \nha dado como resultado que {evento} veces da un numero mayor que 7 o un numero par")

probabilidad = evento / veces
print(f"Probabilidad estimada: {probabilidad}")
```

## Ejercicio 2

Una caja contiene 10 bolas blancas, 20 bolas rojas y 30 bolas verdes. Si tomamos 5 bolas de la caja reemplazándolas (tomando una bola, anotando el color y luego regresandola a la caja). Queremos saber la probabilidad de:

1. Tomar 3 blancas y 2 rojas.
2. Toas todas del mismo color.

Ejecuta el experimento 1000 veces y calcula las probabilidades que mencionadas.

```python

ball_box = {}

# Crea la cajas con las bolas
for i in range(60):
    if i < 10:
        ball_box[i] = "White"
    elif (i > 9) and (i < 30):
        ball_box[i] = "Red"
    else:
        ball_box[i] = "Green"

print(ball_box)

def sacar_bola():
    return ball_box[random.randint(0,59)]

def experimento_3b_2r(veces,nro):
    evento = 0
    for i in range(veces):
        b = 0
        r = 0
        for j in range(nro):
            color = sacar_bola()
            if color == 'White':
                b += 1
            elif color == 'Red':
                r += 1
        if b == 3 and r == 2:
            evento += 1
    return evento

def experimento_mismo_color(veces,nro):
    evento = 0
    for i in range(veces):
        b = 0
        r = 0
        v = 0
        for j in range(nro):
            color = sacar_bola()
            if color == 'White':
                b += 1
            elif color == 'Red':
                r += 1
            elif color == 'Green':
                v += 1
        if b == 5 or r == 5 or v == 5:
            evento += 1
    return evento

veces = 1000
nro = 5
prob_evento1 = experimento_3b_2r(veces,nro)/veces
prob_evento2 = experimento_mismo_color(veces,nro)/veces
print(f"La probabilidad de sacar 3 blancas y 2 rojas es: {prob_evento1}.")
print(f"Mientras que la probabilidad de sacar todas del mismo color es: {prob_evento2}.")

```

## ¿Qué es combinatoria?

La combinatoria es una **forma de contar** cuántas maneras distintas hay de **elegir elementos de un conjunto cuando el orden NO importa**.

Es decir, si eliges los mismos elementos pero en distinto orden, **se considera la misma combinación**.

Se calcula con la siguiente fórmula:

$$
C(n,k) = \frac{n!}{k!(n-k)!}
$$

Donde:

- **n** es el número total de elementos disponibles.
- **k** es el número de elementos que se eligen.
- **!** indica el factorial de un número.

En combinatoria, lo importante es **qué elementos se eligen**, no el orden en el que aparecen.

### Ejemplo de combinatoria

Supongamos que tenemos tres letras: A, B y C, y queremos elegir **dos** de ellas.

Las posibles selecciones son:
- A y B
- A y C
- B y C

Elegir A y B es lo mismo que elegir B y A, por lo que **el orden no importa**.  
En este caso solo existen **3 combinaciones**.

## ¿Qué es permutación?

La permutación es una **forma de contar** cuántas maneras distintas hay de **ordenar o acomodar elementos cuando el orden SÍ importa**.

Es decir, aunque se usen los mismos elementos, **si el orden cambia, cuenta como un resultado diferente**.

Se calcula con la siguiente fórmula:

$$
P(n,k) = \frac{n!}{(n-k)!}
$$

Donde:

- **n** es el número total de elementos disponibles.
- **k** es el número de elementos que se ordenan.
- **!** indica el factorial de un número.

En permutación, lo importante no es solo qué elementos se eligen, sino **en qué orden aparecen**.

### Ejemplo de permutación

Supongamos que tenemos las mismas tres letras: A, B y C, y ahora queremos **ordenar dos** de ellas.

Las posibles opciones son:
- AB
- BA
- AC
- CA
- BC
- CB

En este caso, AB y BA **no son lo mismo**, porque el orden cambia el resultado.  
Por lo tanto, existen **6 permutaciones**.

### En resumen

Cuando el orden de los elementos no importa se utiliza combinatoria, y cuando el orden sí importa se utilizan permutaciones.

## Relación entre combinatoria, permutación y probabilidad

En probabilidad, muchas veces necesitamos saber **cuántos resultados posibles existen** y **cuántos de esos resultados cumplen una condición**.

Aquí es donde entran la combinatoria y la permutación.

- **La combinatoria** se utiliza cuando queremos contar resultados en los que **el orden no importa**.
- **La permutación** se utiliza cuando queremos contar resultados en los que **el orden sí importa**.

Una vez sabemos cuántos resultados posibles hay y cuántos son favorables, podemos calcular la probabilidad de un evento con la siguiente fórmula:

$$
P(\text{evento}) = \frac{\text{número de casos favorables}}{\text{número de casos posibles}}
$$

De esta forma, la combinatoria y la permutación nos ayudan a contar correctamente los casos necesarios para calcular probabilidades.

Por ejemplo, al calcular la probabilidad de obtener ciertas cartas en una mano, usamos combinatoria si el orden de las cartas no importa, y permutación si el orden en el que salen sí es relevante.

## ¿Cuántas alineaciones de 11 jugadores puede hacer?

Como el **orden de los jugadores no importa** (una alineación es la misma sin importar cómo numeremos a los jugadores), estamos ante un **problema de combinatoria**.

La fórmula de la combinatoria es:

$$
\binom{n}{k} = \frac{n!}{k!(n-k)!}
$$

Donde:

- $n$ es el número total de elementos disponibles (jugadores disponibles).
- $k$ es el número de elementos que se eligen (jugadores que forman el equipo).

En este caso:

- $n = 22$
- $k = 11$

Sustituyendo en la fórmula:

$$
\binom{22}{11} = \frac{22!}{11! \cdot 11!}
$$

El resultado es:

$$
\binom{22}{11} = 705\,432
$$

Por lo tanto, el entrenador puede formar **705 432 alineaciones diferentes** de 11 jugadores.

## ¿Cuántas opciones tienes si debes escoger tres asignaturas entre seis optativas?

Como el **orden no importa** (el conjunto de asignaturas es el mismo sin importar el orden en que se elijan), se trata de un **problema de combinatoria**.

La fórmula de la combinatoria es:

$$
\binom{n}{k} = \frac{n!}{k!(n-k)!}
$$

Donde:

- $n$ es el número total de asignaturas disponibles.
- $k$ es el número de asignaturas que se eligen.

En este caso:

- $n = 6$
- $k = 3$

Sustituyendo en la fórmula:

$$
\binom{6}{3} = \frac{6!}{3! \cdot 3!}
$$

Calculando:

$$
\binom{6}{3} = 20
$$

Por lo tanto, existen **20 combinaciones distintas** de tres asignaturas que se pueden elegir entre seis optativas.

```python
# RESPUESTA CON PYTHON + Combinatoria de Math

import math

n_optativas = 6
n_elegir = 3
opciones = math.comb(n_optativas, n_elegir)
opciones
```

## ¿De cuántas formas se pueden sentar tres personas en seis sillas?

En este caso **el orden sí importa**, porque no es lo mismo que una persona se siente en una silla u otra.
Por eso **no usamos combinatoria**, sino **permutación**.

La fórmula de la permutación es:

$$
P(n, k) = \frac{n!}{(n-k)!}
$$

Donde:
- **n** es el número total de opciones disponibles (sillas)
- **k** es el número de elementos que se colocan (personas)

En este ejercicio:
- $$ n = 6 $$
- $$ k = 3 $$

Sustituimos en la fórmula:

$$
P(6,3) = \frac{6!}{(6-3)!}
$$

Simplificando:

$$
P(6,3) = \frac{6!}{3!}
$$

$$
P(6,3) = 6 \cdot 5 \cdot 4
$$

$$
P(6,3) = 120
$$

Por lo tanto, **tres personas se pueden sentar en seis sillas de 120 formas distintas**.

```python
# RESPUESTA CON PYTHON + permutaciones de Math

sillas = 6
personas = 3
formas = math.perm(sillas, personas)
formas
```

## ¿Cuántas jugadas diferentes se pueden obtener si se sacan cinco cartas de una baraja de 40 cartas?

En este caso **el orden de las cartas no importa**, ya que una jugada está formada por las mismas cartas
independientemente del orden en el que se saquen.

Por lo tanto, se trata de un **problema de combinatoria**.

La fórmula de la combinatoria es:

$$
C(n,k) = \frac{n!}{k!(n-k)!}
$$

Donde:
- **n** es el número total de cartas en la baraja
- **k** es el número de cartas que se sacan

En este ejercicio:
- $$ n = 40 $$
- $$ k = 5 $$

Sustituimos en la fórmula:

$$
\binom{40}{5} = \frac{40!}{5! \cdot 35!}
$$

Simplificando:

$$
\binom{40}{5} = \frac{40 \cdot 39 \cdot 38 \cdot 37 \cdot 36}{5 \cdot 4 \cdot 3 \cdot 2 \cdot 1}
$$

El resultado es:

$$
\binom{40}{5} = 658\,008
$$

Por lo tanto, se pueden obtener **658 008 jugadas diferentes** al sacar cinco cartas de una baraja de 40 cartas.

```python
# RESPUESTA CON PYTHON + Combinatoria de Math

import math

n = 40
k = 5
opciones = math.comb(n, k)
opciones
```

## ¿De cuántas formas se pueden sentar cinco personas en una fila de butacas de un cine?

En este caso, se asume que hay **una butaca para cada persona**, es decir, cinco butacas.
El **orden en el que se sientan las personas sí importa**, ya que cambiar a dos personas de sitio
da lugar a una disposición diferente.

Por lo tanto, se trata de un **problema de permutación**.

La fórmula general de la permutación es:

$$
P(n,k) = \frac{n!}{(n-k)!}
$$

En este ejercicio:
- $$ n = 5 $$
- $$ k = 5 $$

Sustituimos en la fórmula:

$$
P(5,5) = \frac{5!}{(5-5)!}
$$

Como \(5 - 5 = 0\) y **por definición \(0! = 1\)**, la expresión se simplifica a:

$$
P(5,5) = \frac{5!}{1} = 5!
$$

Calculando el factorial:

$$
5! = 5 \cdot 4 \cdot 3 \cdot 2 \cdot 1 = 120
$$

Por lo tanto, **cinco personas se pueden sentar en una fila de butacas de 120 formas diferentes**.

```python
# RESPUESTA CON PYTHON + permutaciones de Math

sillas = 5
personas = 5
formas = math.perm(sillas, personas)
formas
```
